### Feature Engineering

```python
# Age Groups
df['AgeGroup'] = pd.cut( 
    df['Age'], 
    bins=[0,40,55,70,100], 
    labels=['Young','Middle','Senior','Old']
)
```

```python
df.sample(5)
```

```python
# Blood Pressure Category
df['BP_Category'] = pd.cut(
    df['RestingBP'], 
    bins=[0,120,140,180,300], 
    labels=['Normal','Elevated','High','VeryHigh']
)
```

```python
df.sample(5)
```

```python
# Cholesterol Category 
df['Chol_Category'] = pd.cut(
    df['Cholesterol'], 
    bins=[0,200,240,700], 
    labels=['Normal','Borderline','High']
)
```

```python
df.sample(5)
```

```python
# Age × Blood Pressure
df['Age_BP'] = df['Age'] * df['RestingBP']
```

```python
df.sample(5)
```

```python
# Age × MaxHR 
df['Age_MaxHR'] = df['Age'] * df['MaxHR']
```

```python
df.sample(5)
```

```python
# Cholesterol/BP Ratio 
df['Chol_BP_Ratio'] = df['Cholesterol'] / (df['RestingBP'] + 1)
```

```python
df.sample(5)
```

```python
# Heart Rate Ratio
df['HR_Age_Ratio'] = df['MaxHR'] / df['Age']
```

```python
df.sample(5)
```
